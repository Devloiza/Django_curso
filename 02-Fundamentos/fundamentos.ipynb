# Funciones de orden superior (HOF)
Son funciones que reciben otras funciones como argumento

```python
# HOF
def require_auth(func):
    def wrapper(user):
        if user.lower() == "admin":
            return func(user)
        else:
            return "Acceso denegado"
    
    return wrapper

# Función a evaluar
def admin_dashboard(user):
    return f"Bienvenido al panel, {user}"

# Probamos la función
auth_view_dashboard = require_auth(admin_dashboard)
print(auth_view_dashboard("Admin"))
print(auth_view_dashboard("Invitado"))
```

# Decoradores

Recibe una función y devuelve una función.

```python
def require_auth(func):
    def wrapper(user):
        if user.lower() == "admin":
            return func(user)
        else:
            return "Acceso denegado"
    
    return wrapper

@require_auth # Este es el decorador
def admin_dashboard(user):
    return f"Bienvenido al panel, {user}"

# Probamos la función
# auth_view_dashboard = require_auth(admin_dashboard) 
# Ya no es necesario guardarlo como variable.
print(admin_dashboard("Admin"))
print(admin_dashboard("Invitado"))
```

# Atibutos y métodos

```python
class Person:
    species = "Humano"
    def __init__(self, name, age):
        # Atributos de instancia
        self.name = name
        self.age = age

        # Atributo protegido
        self._energy = 100 # Por convención, no realmente está protegida

        # Atributo privado
        self.__password = "1234"

    # Método público
    def work(self):
        return f"{self.name} está trabajando duro"

    # Método protegido
    def _waste_energy(self, quantity):
        self._energy -= quantity
        return self._energy
    
    # Método privado
    def __generate_password(self):
        return f"$${self.name}{self.age}$$"

person1 = Person("Ricardo", 29) # A esto se le llama instanciar
person2 = Person("Fernando", 16)

## Mostrar atributos públicos
print(person1.work())
print(person2.work())

## Mostrar atributos protegidos
print(person1._waste_energy(10))

## Mostrar atributos privados
try: print(person1.__password)
except: print('No se puede mostrar este atributo')

print(person1._Person__password)
print(person1._Person__generate_password())
```

# Classmethods y Staticmethods

```python
class Person:
    species = "Humano"
    
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Cambio a nivel de instancia
    def change_species(self, new_species):
        self.species = new_species

    # Cambio a nivel clase
    @classmethod
    def change_species_class(cls, new_species):
        cls.species = new_species

    # Se puede usar a nivel de clase o a nivel de objeto
    @staticmethod
    def is_older(age):
        return age >= 18

person1 = Person("Ricardo", 29)

print(person1.species)

person1.change_species("Reptiliano")
print(person1.species)

Person.change_species_class("Mugiwara")

person2 = Person("Monkey D. Luffy", 19)
print(person2.species)

print(Person.is_older(19))
print(person2.is_older(person2.age))
```

# POO

```python
class BankAccount:
    def __init__(self, owner, initial_balance):
        self.owner = owner
        self.__balance = initial_balance # Encapsulación
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Saldo insuficiente o monto inválido")

    def check_balance(self):
        return f"Saldo actual: ${self.__balance}"
    
account = BankAccount("Ricardo", 1000) # Abstracción
account.deposit(500)
account.withraw(700)
print(account.check_balance())
```

```python
from abc import ABC, abstractmethod

## Clase abstracta
class BankAccount(ABC):
    def __init__(self, owner, initial_balance):
        self.owner = owner
        self.__balance = initial_balance # Encapsulación
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
    
    def _get_balance(self):
        return self.__balance
    
    def _set_balance(self, new_balance):
        self.__balance = new_balance

    @abstractmethod
    def withraw(self, amount):
        pass # Polimorfismo

    def check_balance(self):
        return f"Saldo actual: ${self.__balance}"
    
# Ambas clases heradan todo lo de bankaccount pero hacen su propio método de withraw
class SavingAccount(BankAccount):
    def withraw(self, amount):
        penalty = amount*0.05
        total = amount+penalty
        if total <= self._get_balance():
            self._set_balance(self._get_balance() - total)
        else:
            print("Fondos insuficientes en la cuenta de ahorro")

class PayrollAccount(BankAccount):
    def withraw(self, amount):
        if amount <= self._get_balance():
            self._set_balance(self._get_balance() - amount)
        else:
            print("Fondos insuficientes en la cuenta de ahorro")
        

savings = SavingAccount("Ricardo", 1000)
payroll = PayrollAccount("Ricardo", 1000)

savings.withraw(100)
payroll.withraw(100)

print(savings.check_balance())
print(payroll.check_balance())
```

# Manejo de errores

```python
def divide_numbers():
    try:
        a = int(input("Ingresa el numerador: "))
        b = int(input("Ingresa el denominador: "))
        result = a/b
        return result
    except ZeroDivisionError:
        print("División entre 0")
    except ValueError:
        print("Por favor ingresa sólo números")
    except Exception as e:
        print(type(e))
    else:
        pass
    finally:
        print("Gracias por usarlo")
    

divide_numbers()
```

# Manejo de archivos

```python
try:
    # Modo escritura
    with open("test.txt", mode="w") as my_file:
        text = my_file.write(":) ")

    # Modo lectura
    with open('test.txt', mode="r") as my_file:
        print(my_file.readlines())
    
    # Modo escritura y lectura
    with open('test.txt', mode="r+") as my_file:
        print(my_file.readlines())
        text = my_file.write("Hola mundo ")

    # Modo append
    with open('test.txt', mode="a") as my_file:
        text = my_file.write("123 ")
        print(text)

except FileNotFoundError:
    print("El archivo no existe.")
except Exception as e:
    print(f'Ocurrió un error: {e}')
```
